```python
AIR_TEMP = 'air' # 1.Daily surface air temp. (K)
PRECIP_AMT = 'apcp' # 2.Daily accumulated total surface precip. (kg/m2)
CONV_PRECIP_AMT = 'acpcp' # 3.Daily accumulated conv. surface precip. (kg/m2)
TOTAL_CLOUD_COVER = 'tcdc' # 4.Daily forecast total cloud cover. (%)
UWND = 'uwnd' # 5.Daily u-wind eastward. (m/s)
DEW_POINT_TEMP = 'dpt' # 6.Daily dew point temp. (K)
ALBEDO = 'albedo' # 7.Daily surface albedo. (%)
POT_TEMP = 'pottmp' # 8.Daily potential surface temp. (K)
VWND = 'vwnd' # 9.Daily v-wind nothward. (m/s)
RH = 'rhum' # 10.Daily RH. (%)

PLANT_CODE = 'plant_code'
LATITUDE = 'lat'
LONGITUDE = 'lon'
MONTH = 'month'
YEAR = 'year'
DAY = 'day'

MAX = '_max'
MIN = '_min'
AVG = '_avg'
```

#### NARR var-agg-monthly dfs

```python
# var_df_dict = {}
for var_ in var_names:
    print(var_, '##############')
#     var_df_dict[var_] = {}
    for yr in years_str_list:
        var_df = df({PLANT_CODE: [], YEAR: [], MONTH: [],
                     var_ + MIN: [], var_ + MAX: [], var_ + AVG: []})
        var_final_dfs = [file for file in narr_final_dfs if var_ + '.' + yr in file]
        print(var_final_dfs[0], len(var_final_dfs))
        for var_file in var_final_dfs:
            _, year, month, _ = var_file.split('.')
            daily_var_df = pd.read_excel(
                NARR_FINAL_DATA_MONTHLY_FOLDER + var_file)
            if (var_ == 'uwnd' or var_ == 'vwnd'):
                daily_var_df[var_] = np.abs(daily_var_df[var_])
            monthly_min_df = daily_var_df[[PLANT_CODE, var_]].groupby(PLANT_CODE).min()
            monthly_min_df.reset_index(inplace=True)

            monthly_max_df = daily_var_df[[PLANT_CODE, var_]].groupby(PLANT_CODE).max()
            monthly_max_df.reset_index(inplace=True)

            monthly_avg_df = daily_var_df[[PLANT_CODE, var_]].groupby(PLANT_CODE).mean()
            monthly_avg_df.reset_index(inplace=True)

            var_df[PLANT_CODE] = monthly_avg_df[PLANT_CODE]
            var_df[YEAR] = int(year)
            var_df[MONTH] = int(month)        
            var_df[var_ + MIN] = monthly_min_df[var_]
            var_df[var_ + MAX] = monthly_max_df[var_]
            var_df[var_ + AVG] = monthly_avg_df[var_]
    #         var_df = pd.concat([var_df, df(append_dict)])
            var_df.reset_index(inplace=True, drop=True)
            print(var_df.shape, var_file)
            filename = var_ + '.agg.' + year + '.' + month
            var_df.to_excel(NARR_FINAL_DATA_AGG_FOLDER + filename +'.xlsx', index=False)    


    

```

#### NARR - var-wise final aggregated xlsx [plant_code, year, month, var_min, var_max, var_avg]

```python
# var_df_dict = {}
for var_ in var_names[1:]:
    var_df = df({PLANT_CODE: [], YEAR: [], MONTH: [],
                 var_ + MIN: [], var_ + MAX: [], var_ + AVG: []})
    print(var_, '##############')
    for yr in years_str_list:
        var_final_dfs = [file for file in narr_final_dfs if var_ + '.' + yr in file]
        print(var_final_dfs[0], len(var_final_dfs))
        for var_file in var_final_dfs:
            _, year, month, _ = var_file.split('.')
            daily_var_df = pd.read_excel(
                NARR_FINAL_DATA_MONTHLY_FOLDER + var_file)
            if (var_ == 'uwnd' or var_ == 'vwnd'):
                daily_var_df[var_] = np.abs(daily_var_df[var_])
            monthly_min_df = daily_var_df[[PLANT_CODE, var_]].groupby(PLANT_CODE).min()
            monthly_min_df.reset_index(inplace=True)

            monthly_max_df = daily_var_df[[PLANT_CODE, var_]].groupby(PLANT_CODE).max()
            monthly_max_df.reset_index(inplace=True)

            monthly_avg_df = daily_var_df[[PLANT_CODE, var_]].groupby(PLANT_CODE).mean()
            monthly_avg_df.reset_index(inplace=True)
            
            append_dict = {PLANT_CODE: monthly_avg_df[PLANT_CODE],
                           YEAR: [int(year) for ent in monthly_avg_df[PLANT_CODE]],
                           MONTH: [int(month) for ent in monthly_avg_df[PLANT_CODE]],
                           var_+ MIN: monthly_min_df[var_],
                           var_+ MAX: monthly_max_df[var_],
                           var_+ AVG: monthly_avg_df[var_]}

#             var_df[PLANT_CODE] = monthly_avg_df[PLANT_CODE]
#             var_df[YEAR] = int(year)
#             var_df[MONTH] = int(month)        
#             var_df[var_ + MIN] = monthly_min_df[var_]
#             var_df[var_ + MAX] = monthly_max_df[var_]
#             var_df[var_ + AVG] = monthly_avg_df[var_]
            var_df = pd.concat([var_df, df(append_dict)])
            var_df.reset_index(inplace=True, drop=True)
            print(var_df.shape, var_file)
    filename = var_ + '.agg.' + str(years_str_list[0]) + '.' + str(years_str_list[-1])
    var_df.to_excel(NARR_FINAL_DATA_VAR_AGG_FOLDER + filename +'.xlsx', index=False)    


    

```
