```python
DATA_STORE_PATH = []
var_data_list = []
time_date_data_list = []
lat_data_list = []
lon_data_list = []

for ncfile, VAR_NAME in zip(NCFILE, VARIABLE_NAME):
    DATA_STORE_PATH.append(NARR_FINAL_DATA_FOLDER + VAR_NAME + '/')
    temp = ncfile.variables[VAR_NAME]
    var_data = temp[:]*1
    print(VAR_NAME, var_data.shape)
    var_data_stack = np.concatenate(var_data, axis=None)
    var_data_list.append(var_data)
    print(var_data_stack.shape)

    temp = ncfile.variables[TIME]
    time_data = temp[:]*1
    time_date_data = [REF_DATE + datetime.timedelta(hours=ts) for ts in time_data]
    time_date_data = [date_time.date() for date_time in time_date_data]
    time_date_data_list.append(time_date_data)
    print(TIME, time_data.shape, len(time_date_data))

    temp = ncfile.variables[LATITUDE]
    lat_data = temp[:]*1
    lat_data_stack = np.concatenate(lat_data, axis=None)
    print(lat_data_stack[0:10], lat_data_stack.shape)
    lat_data_stack = np.round(lat_data_stack, 2)
    print(lat_data_stack[0:10], lat_data_stack.shape)
    lat_data_list.append(lat_data_stack)

    temp = ncfile.variables[LONGITUDE]
    lon_data = temp[:]*1
    lon_data_stack = np.concatenate(lon_data, axis=None)
    print(lon_data_stack[0:10], lon_data_stack.shape)
    lon_data_stack = np.round(lon_data_stack, 2)
    print(lon_data_stack[0:10], lon_data_stack.shape)
    lon_data_list.append(lon_data_stack)

    print(LATITUDE, lat_data.shape)
    print(LONGITUDE, lon_data.shape)

print(len(var_data_list), len(time_date_data_list), len(lat_data_list), len(lon_data_list))
```
